### Aim:
To demonstrate  Clustering of Data using K-means Clustering Technique for Data Science applications

### Description:
K-means clustering algorithm computes the centroids and iterates until we it finds optimal centroid. The number of clusters identified from data by algorithm is
represented by ‘K’ in K-means. In this algorithm, the data points are assigned to a cluster in such a manner that the sum of the squared distance between the data
points and centroid would be minimum. It is to be understood that less variation within the clusters will lead to more similar data points within same cluster

#### Develop a K-means clustering model for the Mall_Customers dataset using the scikit-learn

##### a. Use the columns: ' Age', 'Annual_Income' as the input variables.
##### b. Compute the optimal number of cluster ‘K’ from 1-10 using the Elbow method.
##### c. Plot the graph between number of cluster K and within-cluster sum of squares (WCSS) value.
##### d. Perform the K-means clustering with the selected optimal K.
##### e. Display the cluster centroids.
##### f. Visualize the data representation of K-means clustering.
##### g. Change the value of K in K-means with different values and tabulate the performance metrics such as silhouette_score and davies_bouldin_score obtained.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
data = pd.read_csv('Mall_Customers.csv')
X = data[['Age', 'Annual Income (k$)']]
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

print("URK22CS5094\n")
plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Within-cluster sum of squares (WCSS)')
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=0)
clusters = kmeans.fit_predict(X)
data['Cluster'] = clusters
```

```python
centroids = kmeans.cluster_centers_
print ("URK22CS5094\n")
print("Cluster Centroids:\n", centroids)
```

```python
plt.scatter(X['Age'], X['Annual Income (k$)'], c=clusters, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids')
print ("URK22CS5094\n")
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
plt.show()
```

```python
results = []

for k in range(optimal_k, optimal_k + 4):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  
    clusters = kmeans.fit_predict(X)
    
    silhouette_avg = silhouette_score(X, clusters)
    db_score = davies_bouldin_score(X, clusters)
    
    results.append((k, silhouette_avg, db_score))

results_df = pd.DataFrame(results, columns=['K-value', 'Silhouette Score', 'Davies-Bouldin Score'])

print ("URK22CS5094\n")
print(results_df)
```
